# Dataframe de coordenadas de estaciones

No tenemos un dataset de coordenadas de estaciones de bicicleta, sólo registros de viajes. Tenemos que extraer el id y las coordenadas de cada df

```python
# Primero se buscan todas las estaciones de bicicleta

import pandas as pd

df=pd.read_csv('./Datos/datos_sucios/2013-07 - Citi Bike trip data.csv')
df.sample(5)
```

```python
import os

# Lista de los archivos en datos_sucios
csvs=os.listdir('./Datos/datos_sucios') 
csvs=[f'./Datos/datos_sucios/{csv}' for csv in csvs]
csvs
```

```python
import pandas as pd

df_todos=pd.read_csv(csvs[0])

for archivo in csvs[1:]:
  df=pd.read_csv(archivo)
  df_todos=df_todos.append(df)

df_todos
```

```python
df_todos.to_csv('./Datos/datos_limpios/datos_completos.csv',index=False)
```

```python
pd.read_csv('./Datos/datos_limpios/datos_completos.csv')
```

# Ids con nombre de estación y coordenadas


```python
df_estacion=df_todos[['start station id','start station name','start station latitude','start station longitude']].drop_duplicates()
df_estacion
```

```python
df_estacion.to_csv('./Datos/datos_limpios/datos_estacion.csv')
```
